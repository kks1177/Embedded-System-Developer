# 이미지 구분 AI 모델링_Segmentation

```python
from bready.semantic_segmentation_tools import SemanticSegmentation
```

```python
trainer = SemanticSegmentation.Trainer()
```

```python
train_data_path = "C:/Users/user/AI_Jupyter_Example/03_Segmentation/seg1_train"
validation_data_path = "C:/Users/user/AI_Jupyter_Example/03_Segmentation/seg1_validation"
```

```python
trainer.setTrainingDirectory(train_data_path)
trainer.setValidationDirectory(validation_data_path)
```

```python
save_model_path = "training_models"
trainer.setSaveModelDirectory(save_model_path)
```

```python
learning_rate = 0.0001
epochs = 40
batch_size = 4
```

```python
trainer.setLearningRate(learning_rate)
trainer.setEpochs(epochs)
trainer.setBatchSize(batch_size)
```

```python
trainer.setOptimizer(SemanticSegmentation.Keras_Optimizer.ADAM)
```

```python
trainer.setAugmentation_Option(SemanticSegmentation.Image_Augmentation.HORIZONTAL_FLIP, True)
trainer.setAugmentation_Option(SemanticSegmentation.Image_Augmentation.VERTICAL_FLIP, True)
trainer.setAugmentation_Option(SemanticSegmentation.Image_Augmentation.RANDOM_BRIGHTNESS, True)
trainer.setAugmentation_Option(SemanticSegmentation.Image_Augmentation.RANDOM_CROP, True)
```

```python
trainer.setModelSaveType(SemanticSegmentation.ModelSaveType.SMALLEST_ERROR)
```

```python
trainer.train()
```

```python
trainer.dispose()
```
