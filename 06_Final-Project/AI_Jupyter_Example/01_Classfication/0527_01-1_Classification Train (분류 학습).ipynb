# 이미지 분류 AI 모델링_Classification

```python
from bready.classification_tools import Classifier
```

```python
trainer = Classifier.Trainer()
```

```python
train_data_path = "C:/Embedded_System/06_Raspberry Pi/05_비전(시각형) AI/AI-X/classification1_dataset/class1_train"
validation_data_path = "C:/Embedded_System/06_Raspberry Pi/05_비전(시각형) AI/AI-X/classification1_dataset/class1_validation"
```

```python
trainer.setTrainingDirectory(train_data_path)
trainer.setValidationDirectory(validation_data_path)
```

```python
save_model_path = "training_models"
trainer.setSaveModelDirectory(save_model_path)
```

```python
learning_rate = 0.0001
epochs = 10
batch_size = 8
```

```python
trainer.setLearningRate(learning_rate)
trainer.setEpochs(epochs)
trainer.setBatchSize(batch_size)
```

```python
trainer.setOptimizer(Classifier.Keras_Optimizer.ADAM_OPTIMIZER)
```

```python
trainer.setAugmentation_Option(Classifier.Image_Augmentation.HORIZONTAL_FLIP, True)
```

```python
trainer.setModelSaveType(Classifier.ModelSaveType.SMART_OPTION)
```

```python
trainer.train()
```

```python
trainer.dispose()
```
